```python
import decimal
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy

# Начальные условия

xn_array = []
x0 = 0
dx0 = 0.001
T0 = 0
Tk = 1000 
mu = 1000

N = 1000000
h = (Tk - T0)/N # Шаг по времени

def Euler_method(xn1, dxn1):
    xn2 = xn1 + h*dxn1   # координата
    dxn2 = dxn1 + h*(mu*(1-dxn1**2)*dxn1 - xn1) # 2 производная
    return (xn2, dxn2)

Tn = T0
xn = 0
dxn = dx0
xn_array = [xn]
while Tn < Tk:
    xn, dxn = Euler_method(xn, dxn)
    xn_array.append(xn)
    Tn += h


plt.plot(np.linspace(0, Tk, len(xn_array)), xn_array)
```
